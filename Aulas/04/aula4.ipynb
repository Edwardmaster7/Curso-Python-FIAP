# List Comprehension
 - List Comprehension é uma forma concisa e eficiente de criar listas em Python a partir de iteráveis, podendo aplicar transformações e filtros de maneira elegante.

```python
def jeito_antigo(numeros:list) -> list:
    final =[]
    for n in numeros:
        if n % 2 == 0:
            final.append(n * n)
    return final

def jeito_novo(numeros:list) -> list:
    return [n * n for n in numeros if n % 2 == 0]
```

```python
lista_numeros = list[int](range(int(input("Digite um número:\n"))))
```

```python
antigo = jeito_antigo(lista_numeros)
print(f'Modo Antigo: {antigo}')
```

```python
antigo = jeito_novo(lista_numeros)
print(f'Modo Novo: {antigo}')
```

> Ex 01:

Dada a lista nomes = ['ana', 'bruno', 'carla'], use uma
list comprehension para criar duas novas listas:

- Uma onde cada nome esteja em letras maiúsculas.
- Outra em CammelCase

== nomes = ['ana', 'bruno', 'carla']

```python
nomes:list[str] = ["ana", "bruno", "carla"]

caps_list = [nome.upper() for nome in nomes]
cc_list = [nome.capitalize() for nome in nomes]

print(f'Capitalized: {caps_list}')
print(f'CamelCased: {cc_list}')
```

> Ex 02:

Dada a lista numeros = , use uma list comprehension para 
criar uma nova lista contendo apenas os números que são maiores que 8.


== numeros = [5,12,67,6,4,9,8,21,7,3,45,2]

```python
numeros = [5,12,67,6,4,9,8,21,7,3,45,2]

bigger_than_eight = [numero for numero in numeros if numero > 8]

print(f'Maiores que 8: {bigger_than_eight}')
```
