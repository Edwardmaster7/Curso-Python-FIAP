## Gerenciamento de contexto com `with`

```python

with open('file.txt') as f:
    arq = f.read()
```

```python
print(arq) 
```

```python
class Leitura:
    def __init__(self, nome:str) -> None:
        self.nome = nome
    
    def __enter__(self):
        print('Hahaha')
    
    def __exit__(self):
        print('saiu')
    
    def __repr__(self):
        return f'Leitura({self.nome})'
```

```python
Leitura('file.txt')
```

## Dict example

```python
dicionario = {}
```

```python
dicionario['chave'] = 'valor'
```

```python
print(dicionario)
```

```python
dicionario[0] = 'teste'
print(dicionario)
```

```python
dict_cars = {}
empty_positions = list[int](range(1,3))
complete_positions = []
```

```python
print(dict_cars)
print(empty_positions)
print(complete_positions)
```

```python
dict_cars[empty_positions[0]] = 'Teste'
complete_positions.append(empty_positions[0])
empty_positions.pop(0)
```

```python
print(dict_cars)
print(empty_positions)
print(complete_positions)
```

```python
print(dict_cars.values())
```
