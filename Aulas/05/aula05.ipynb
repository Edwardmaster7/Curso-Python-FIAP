Crie uma função chamada calcular_media que receba uma lista de números (inteiros ou floats) e retorne a média. A função deve ter type hints (anotações de tipo) corretos para os argumentos e para o valor de retorno.

Bônus: Se a lista estiver vazia, a função deve retornar 0.0 (float).

```python
def calcular_media(numeros:list[int]) -> float:
    if not bool(numeros): return 0.0
    total:float = 0.0
    return (total+n for n in numeros)/len(numeros)
    
```

```python
test_list = [1,1,1]
calcular_media(test_list)
```

```python
'''
 ________
|> EX02 <|
Crie uma função:

>>> classificar_imc(peso: float, altura: float) -> str. 

A função deve primeiro calcular o IMC (Índice de Massa Corporal) usando a fórmula:

>>> $peso / (altura * altura)$

Em seguida, deve retornar uma string classificando o resultado:

- Menor que 18.5: "Abaixo do peso"
- 18.5 a 24.9: "Peso normal"
- 25.0 a 29.9: "Sobrepeso"
- 30.0 ou mais: "Obesidade"
'''

def classificar_imc(
        peso:float,
        altura:float
        ) -> str:
    num_imc = peso / (altura * altura)
    
    if num_imc < 18.5: return 'Abaixo do peso'
    elif num_imc < 25: return 'Peso normal'
    elif num_imc < 30: return 'Sobrepeso'
    else: return 'Obesidade'
```

```python
classificar_imc(80, 1.50)
```
